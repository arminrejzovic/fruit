```python
import numpy as np
```

```python
inputs = [[0.54, 0.31, 0.33, 0.62], [0.41, 0.62, 0.12, 0.2], [0.49, 0.64, 0.77, 0.22]] # 4*3
weights = [[0.82, 0.65, 0.68, 0.99], [0.90, 0.30, 0.65, 0.83], [0.66, 0.52, 0.76, 0.76]] # 4*3
biases = [0.54, 0.31, 0.73]

inputs = np.array(inputs)
weights = np.array(weights).T
# Doing transpose to make it 3x4 since you cannot multiply 4x3 with 4x3
# The resulting matrix is 3x3

results = np.dot(inputs, weights) + biases
```

```python
weights2 = [[0.03, 0.34, 0.54, 0.05], [0.04, 0.15, 0.61, 0.62], [0.43, 0.75, 0.07, 0.86]]
biases2 = [[0.48, 0.93, 0.77, 0.55], [0.76, 0.11, 0.05, 0.20], [0.88, 0.51, 0.46, 0.64]]

weights2 = np.array(weights2)
result2 = np.dot(results, weights2) + biases2

result2
```
